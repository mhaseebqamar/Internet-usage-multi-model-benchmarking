```python
# ============================================================================
# INTERNET USAGE DATA - COMPLETE DATA MINING PROJECT
# Dataset: UCI Internet Usage Data (1997)
# Models: Logistic Regression, Decision Tree, KNN, SVM, Neural Network
# ============================================================================

# INSTALLATION (Run this first if needed)
# !pip install scikit-learn pandas numpy matplotlib seaborn

import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import gzip

# Machine Learning Libraries
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                            f1_score, confusion_matrix, classification_report,
                            roc_curve, auc, roc_auc_score)

# Set style for better visualizations
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("=" * 80)
print("INTERNET USAGE DATA MINING PROJECT")
print("=" * 80)

# ============================================================================
# 1. UNDERSTAND DATA - Data Loading and Initial Exploration
# ============================================================================
print("\n" + "=" * 80)
print("STEP 1: DATA UNDERSTANDING")
print("=" * 80)

# Load the dataset from local file
file_path = '/Users/Haseeb/Downloads/final_general.dat.gz'

print(f"\n📂 Loading dataset from: {file_path}")

# Read the gzipped data file (space-separated values)
# Using on_bad_lines='skip' to handle inconsistent columns
try:
    with gzip.open(file_path, 'rt') as f:
        df = pd.read_csv(f, sep='\s+', header=None, on_bad_lines='skip', engine='python')
    print("✅ Dataset loaded successfully!")
except Exception as e:
    print(f"❌ Error loading with python engine, trying alternative method...")
    # Alternative: read line by line and handle manually
    with gzip.open(file_path, 'rt') as f:
        lines = []
        for i, line in enumerate(f):
            parts = line.strip().split()
            if len(parts) == 72:  # Keep only rows with exactly 72 columns
                lines.append(parts)
        df = pd.DataFrame(lines)
        # Convert to numeric where possible
        df = df.apply(pd.to_numeric, errors='ignore')
    print("✅ Dataset loaded successfully using alternative method!")

print(f"   - Loaded {len(df)} valid records")

print(f"\n📏 Dataset Shape: {df.shape}")
print(f"   - Number of Records: {df.shape[0]:,}")
print(f"   - Number of Features: {df.shape[1]}")

print("\n🔍 First 5 Rows:")
print(df.head())

print("\n🔍 Last 5 Rows:")
print(df.tail())

print("\n📊 Data Types:")
print(df.dtypes.value_counts())

print("\n🔢 Descriptive Statistics:")
print(df.describe())

print("\n❓ Missing Values:")
missing = df.isnull().sum()
print(missing[missing > 0] if missing.sum() > 0 else "No missing values found!")

# Check for any placeholder values (like 99, 999, -1 that might represent missing data)
print("\n🔍 Checking for placeholder values...")
for col in df.columns[:10]:  # Check first 10 columns
    unique_vals = df[col].unique()
    if len(unique_vals) < 20:
        print(f"   Column {col}: {sorted(unique_vals)}")

# Visualization 1: Data Overview
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Step 1: Data Understanding - Overview', fontsize=16, fontweight='bold')

# Plot 1: Dataset size
axes[0, 0].bar(['Records', 'Features'], [df.shape[0], df.shape[1]], color=['#3498db', '#e74c3c'])
axes[0, 0].set_title('Dataset Dimensions')
axes[0, 0].set_ylabel('Count')
for i, v in enumerate([df.shape[0], df.shape[1]]):
    axes[0, 0].text(i, v + 50, str(v), ha='center', fontweight='bold')

# Plot 2: Missing values heatmap
missing_matrix = df.isnull().astype(int)
if missing_matrix.sum().sum() > 0:
    sns.heatmap(missing_matrix.iloc[:100, :20], cbar=False, ax=axes[0, 1], cmap='YlOrRd')
    axes[0, 1].set_title('Missing Values Pattern (First 100 rows, 20 cols)')
else:
    axes[0, 1].text(0.5, 0.5, 'No Missing Values!', ha='center', va='center',
                    fontsize=14, fontweight='bold', transform=axes[0, 1].transAxes)
    axes[0, 1].set_title('Missing Values Check')
    axes[0, 1].axis('off')

# Plot 3: Data types distribution
dtype_counts = df.dtypes.astype(str).value_counts()
axes[1, 0].pie(dtype_counts.values, labels=dtype_counts.index, autopct='%1.1f%%', startangle=90)
axes[1, 0].set_title('Feature Data Types Distribution')

# Plot 4: Sample feature distribution
numeric_cols = df.select_dtypes(include=[np.number]).columns[:5]
if len(numeric_cols) > 0:
    df[numeric_cols].boxplot(ax=axes[1, 1])
    axes[1, 1].set_title(f'Distribution of First {len(numeric_cols)} Numeric Features')
    axes[1, 1].set_xticklabels(axes[1, 1].get_xticklabels(), rotation=45, ha='right')
else:
    axes[1, 1].text(0.5, 0.5, 'No numeric features to display', ha='center', va='center',
                    transform=axes[1, 1].transAxes)

plt.tight_layout()
plt.savefig('01_data_understanding.png', dpi=300, bbox_inches='tight')
plt.show()

# ============================================================================
# 2. DATA PREPROCESSING
# ============================================================================
print("\n" + "=" * 80)
print("STEP 2: DATA PREPROCESSING")
print("=" * 80)

# Create a copy for preprocessing
df_processed = df.copy()

# Handle missing values if any
if df_processed.isnull().sum().sum() > 0:
    print("\n🔧 Handling missing values...")
    # For numeric columns: fill with median
    numeric_columns = df_processed.select_dtypes(include=[np.number]).columns
    for col in numeric_columns:
        if df_processed[col].isnull().sum() > 0:
            df_processed[col].fillna(df_processed[col].median(), inplace=True)

    # For categorical columns: fill with mode
    categorical_columns = df_processed.select_dtypes(exclude=[np.number]).columns
    for col in categorical_columns:
        if df_processed[col].isnull().sum() > 0:
            df_processed[col].fillna(df_processed[col].mode()[0], inplace=True)
    print("   ✅ Missing values handled")
else:
    print("\n✅ No missing values to handle")

# Encode categorical variables if any
categorical_cols = df_processed.select_dtypes(exclude=[np.number]).columns
if len(categorical_cols) > 0:
    print(f"\n🔤 Encoding {len(categorical_cols)} categorical columns...")
    le = LabelEncoder()
    for col in categorical_cols:
        df_processed[col] = le.fit_transform(df_processed[col].astype(str))
    print("   ✅ Categorical encoding complete")

# Prepare features and target
# Using the last column as target variable
target_col = df_processed.columns[-1]
X_processed = df_processed.drop(columns=[target_col])
y_processed = df_processed[target_col]

print(f"\n📊 Target Variable Info:")
print(f"   - Unique values: {y_processed.nunique()}")
print(f"   - Value distribution:\n{y_processed.value_counts()}")

# For binary classification, ensure target is binary
if y_processed.nunique() > 2:
    print(f"\n⚠️  Target has {y_processed.nunique()} classes. Converting to binary...")
    # Convert to binary based on median
    y_processed = (y_processed > y_processed.median()).astype(int)
    print(f"   ✅ Converted to binary classification")
    print(f"   - New class distribution:\n{y_processed.value_counts()}")

print(f"\n📊 Processed Dataset:")
print(f"   - Features shape: {X_processed.shape}")
print(f"   - Target shape: {y_processed.shape}")
print(f"   - Target classes: {sorted(y_processed.unique())}")

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X_processed, y_processed, test_size=0.2, random_state=42, stratify=y_processed
)

print(f"\n🔀 Train-Test Split:")
print(f"   - Training set: {X_train.shape[0]:,} samples ({X_train.shape[0]/len(df)*100:.1f}%)")
print(f"   - Test set: {X_test.shape[0]:,} samples ({X_test.shape[0]/len(df)*100:.1f}%)")

# Feature Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
print("\n✅ Feature scaling applied (StandardScaler)")

# Visualization 2: Preprocessing
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Step 2: Data Preprocessing', fontsize=16, fontweight='bold')

# Plot 1: Class distribution
y_processed.value_counts().plot(kind='bar', ax=axes[0, 0], color=['#2ecc71', '#e74c3c'])
axes[0, 0].set_title('Target Variable Distribution')
axes[0, 0].set_xlabel('Class')
axes[0, 0].set_ylabel('Count')
axes[0, 0].set_xticklabels(axes[0, 0].get_xticklabels(), rotation=0)
for i, (idx, val) in enumerate(y_processed.value_counts().items()):
    axes[0, 0].text(i, val + 50, f'{val:,}', ha='center', fontweight='bold')

# Plot 2: Train-Test split
split_data = pd.DataFrame({
    'Dataset': ['Training', 'Test'],
    'Samples': [X_train.shape[0], X_test.shape[0]]
})
axes[0, 1].bar(split_data['Dataset'], split_data['Samples'], color=['#3498db', '#9b59b6'])
axes[0, 1].set_title('Train-Test Split (80-20)')
axes[0, 1].set_ylabel('Number of Samples')
for i, v in enumerate(split_data['Samples']):
    axes[0, 1].text(i, v + 50, f'{v:,}', ha='center', fontweight='bold')

# Plot 3: Feature correlation heatmap (top 10 features)
top_features = X_processed.iloc[:, :10]
correlation = top_features.corr()
sns.heatmap(correlation, annot=True, fmt='.2f', cmap='coolwarm', ax=axes[1, 0],
            square=True, cbar_kws={'shrink': 0.8})
axes[1, 0].set_title('Feature Correlation Matrix (First 10 Features)')

# Plot 4: Feature scaling comparison
feature_idx = 0  # First feature
original_values = X_train.iloc[:100, feature_idx].values
scaled_values = X_train_scaled[:100, feature_idx]

axes[1, 1].plot(original_values, label='Original', alpha=0.7, linewidth=2)
axes[1, 1].plot(scaled_values, label='Scaled', alpha=0.7, linewidth=2)
axes[1, 1].set_title(f'Feature Scaling Effect (Feature: {X_train.columns[feature_idx]})')
axes[1, 1].set_xlabel('Sample Index')
axes[1, 1].set_ylabel('Value')
axes[1, 1].legend()
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('02_data_preprocessing.png', dpi=300, bbox_inches='tight')
plt.show()

# ============================================================================
# 3. LOGISTIC REGRESSION
# ============================================================================
print("\n" + "=" * 80)
print("STEP 3: LOGISTIC REGRESSION")
print("=" * 80)

print("\n🤖 Training Logistic Regression Model...")
lr_model = LogisticRegression(random_state=42, max_iter=1000)
lr_model.fit(X_train_scaled, y_train)

# Predictions
y_pred_lr_train = lr_model.predict(X_train_scaled)
y_pred_lr_test = lr_model.predict(X_test_scaled)
y_pred_lr_proba = lr_model.predict_proba(X_test_scaled)[:, 1]

# Evaluation
lr_train_acc = accuracy_score(y_train, y_pred_lr_train)
lr_test_acc = accuracy_score(y_test, y_pred_lr_test)
lr_precision = precision_score(y_test, y_pred_lr_test, zero_division=0)
lr_recall = recall_score(y_test, y_pred_lr_test, zero_division=0)
lr_f1 = f1_score(y_test, y_pred_lr_test, zero_division=0)

print(f"\n📊 Logistic Regression Results:")
print(f"   - Training Accuracy: {lr_train_acc:.4f} ({lr_train_acc*100:.2f}%)")
print(f"   - Test Accuracy: {lr_test_acc:.4f} ({lr_test_acc*100:.2f}%)")
print(f"   - Precision: {lr_precision:.4f}")
print(f"   - Recall: {lr_recall:.4f}")
print(f"   - F1-Score: {lr_f1:.4f}")

print("\n📋 Classification Report:")
print(classification_report(y_test, y_pred_lr_test))

# Confusion Matrix
cm_lr = confusion_matrix(y_test, y_pred_lr_test)
print("\n🔢 Confusion Matrix:")
print(cm_lr)

# ROC Curve
fpr_lr, tpr_lr, _ = roc_curve(y_test, y_pred_lr_proba)
roc_auc_lr = auc(fpr_lr, tpr_lr)

# Visualization 3: Logistic Regression
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Step 3: Logistic Regression Analysis', fontsize=16, fontweight='bold')

# Plot 1: Confusion Matrix
sns.heatmap(cm_lr, annot=True, fmt='d', cmap='Blues', ax=axes[0, 0],
            xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
axes[0, 0].set_title('Confusion Matrix')
axes[0, 0].set_ylabel('True Label')
axes[0, 0].set_xlabel('Predicted Label')

# Plot 2: ROC Curve
axes[0, 1].plot(fpr_lr, tpr_lr, color='darkorange', lw=2,
                label=f'ROC curve (AUC = {roc_auc_lr:.2f})')
axes[0, 1].plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Classifier')
axes[0, 1].set_xlim([0.0, 1.0])
axes[0, 1].set_ylim([0.0, 1.05])
axes[0, 1].set_xlabel('False Positive Rate')
axes[0, 1].set_ylabel('True Positive Rate')
axes[0, 1].set_title('ROC Curve')
axes[0, 1].legend(loc="lower right")
axes[0, 1].grid(True, alpha=0.3)

# Plot 3: Metrics Comparison
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
scores = [lr_test_acc, lr_precision, lr_recall, lr_f1]
bars = axes[1, 0].bar(metrics, scores, color=['#3498db', '#2ecc71', '#e74c3c', '#f39c12'])
axes[1, 0].set_title('Performance Metrics')
axes[1, 0].set_ylabel('Score')
axes[1, 0].set_ylim([0, 1.1])
for bar in bars:
    height = bar.get_height()
    axes[1, 0].text(bar.get_x() + bar.get_width()/2., height,
                    f'{height:.3f}', ha='center', va='bottom', fontweight='bold')

# Plot 4: Feature Importance (Coefficients)
num_features_to_show = min(10, len(X_train.columns))
feature_importance_lr = pd.DataFrame({
    'Feature': X_train.columns[:num_features_to_show],
    'Coefficient': np.abs(lr_model.coef_[0][:num_features_to_show])
}).sort_values('Coefficient', ascending=True)

axes[1, 1].barh(feature_importance_lr['Feature'], feature_importance_lr['Coefficient'],
                color='steelblue')
axes[1, 1].set_title(f'Top {num_features_to_show} Feature Importance (Absolute Coefficients)')
axes[1, 1].set_xlabel('Absolute Coefficient Value')

plt.tight_layout()
plt.savefig('03_logistic_regression.png', dpi=300, bbox_inches='tight')
plt.show()

# ============================================================================
# 4. DECISION TREE
# ============================================================================
print("\n" + "=" * 80)
print("STEP 4: DECISION TREE")
print("=" * 80)

print("\n🌳 Training Decision Tree Model...")
dt_model = DecisionTreeClassifier(random_state=42, max_depth=10)
dt_model.fit(X_train_scaled, y_train)

# Predictions
y_pred_dt_train = dt_model.predict(X_train_scaled)
y_pred_dt_test = dt_model.predict(X_test_scaled)
y_pred_dt_proba = dt_model.predict_proba(X_test_scaled)[:, 1]

# Evaluation
dt_train_acc = accuracy_score(y_train, y_pred_dt_train)
dt_test_acc = accuracy_score(y_test, y_pred_dt_test)
dt_precision = precision_score(y_test, y_pred_dt_test, zero_division=0)
dt_recall = recall_score(y_test, y_pred_dt_test, zero_division=0)
dt_f1 = f1_score(y_test, y_pred_dt_test, zero_division=0)

print(f"\n📊 Decision Tree Results:")
print(f"   - Training Accuracy: {dt_train_acc:.4f} ({dt_train_acc*100:.2f}%)")
print(f"   - Test Accuracy: {dt_test_acc:.4f} ({dt_test_acc*100:.2f}%)")
print(f"   - Precision: {dt_precision:.4f}")
print(f"   - Recall: {dt_recall:.4f}")
print(f"   - F1-Score: {dt_f1:.4f}")

cm_dt = confusion_matrix(y_test, y_pred_dt_test)
fpr_dt, tpr_dt, _ = roc_curve(y_test, y_pred_dt_proba)
roc_auc_dt = auc(fpr_dt, tpr_dt)

# Visualization 4: Decision Tree
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Step 4: Decision Tree Analysis', fontsize=16, fontweight='bold')

# Plot 1: Confusion Matrix
sns.heatmap(cm_dt, annot=True, fmt='d', cmap='Greens', ax=axes[0, 0],
            xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
axes[0, 0].set_title('Confusion Matrix')
axes[0, 0].set_ylabel('True Label')
axes[0, 0].set_xlabel('Predicted Label')

# Plot 2: ROC Curve
axes[0, 1].plot(fpr_dt, tpr_dt, color='green', lw=2,
                label=f'ROC curve (AUC = {roc_auc_dt:.2f})')
axes[0, 1].plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Classifier')
axes[0, 1].set_xlim([0.0, 1.0])
axes[0, 1].set_ylim([0.0, 1.05])
axes[0, 1].set_xlabel('False Positive Rate')
axes[0, 1].set_ylabel('True Positive Rate')
axes[0, 1].set_title('ROC Curve')
axes[0, 1].legend(loc="lower right")
axes[0, 1].grid(True, alpha=0.3)

# Plot 3: Metrics
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
scores = [dt_test_acc, dt_precision, dt_recall, dt_f1]
bars = axes[1, 0].bar(metrics, scores, color=['#27ae60', '#16a085', '#1abc9c', '#48c9b0'])
axes[1, 0].set_title('Performance Metrics')
axes[1, 0].set_ylabel('Score')
axes[1, 0].set_ylim([0, 1.1])
for bar in bars:
    height = bar.get_height()
    axes[1, 0].text(bar.get_x() + bar.get_width()/2., height,
                    f'{height:.3f}', ha='center', va='bottom', fontweight='bold')

# Plot 4: Feature Importance
num_features_to_show = min(10, len(X_train.columns))
feature_importance_dt = pd.DataFrame({
    'Feature': X_train.columns[:num_features_to_show],
    'Importance': dt_model.feature_importances_[:num_features_to_show]
}).sort_values('Importance', ascending=True)

axes[1, 1].barh(feature_importance_dt['Feature'], feature_importance_dt['Importance'],
                color='forestgreen')
axes[1, 1].set_title(f'Top {num_features_to_show} Feature Importance')
axes[1, 1].set_xlabel('Importance Score')

plt.tight_layout()
plt.savefig('04_decision_tree.png', dpi=300, bbox_inches='tight')
plt.show()

# Visualization 4b: Actual Decision Tree Structure
print("\n🌳 Generating Decision Tree Structure Visualization...")
fig, ax = plt.subplots(figsize=(20, 12))
plot_tree(dt_model,
          feature_names=X_train.columns,
          class_names=['Class 0', 'Class 1'],
          filled=True,
          rounded=True,
          fontsize=10,
          max_depth=3,  # Limit depth for readability
          ax=ax)
ax.set_title('Decision Tree Structure (Max Depth=3 for Visualization)',
             fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('04b_decision_tree_structure.png', dpi=300, bbox_inches='tight')
plt.show()
print("   ✅ Tree structure saved as '04b_decision_tree_structure.png'")

# ============================================================================
# 5. K-NEAREST NEIGHBORS (KNN)
# ============================================================================
print("\n" + "=" * 80)
print("STEP 5: K-NEAREST NEIGHBORS (KNN)")
print("=" * 80)

print("\n👥 Training KNN Model...")
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train_scaled, y_train)

# Predictions
y_pred_knn_train = knn_model.predict(X_train_scaled)
y_pred_knn_test = knn_model.predict(X_test_scaled)
y_pred_knn_proba = knn_model.predict_proba(X_test_scaled)[:, 1]

# Evaluation
knn_train_acc = accuracy_score(y_train, y_pred_knn_train)
knn_test_acc = accuracy_score(y_test, y_pred_knn_test)
knn_precision = precision_score(y_test, y_pred_knn_test, zero_division=0)
knn_recall = recall_score(y_test, y_pred_knn_test, zero_division=0)
knn_f1 = f1_score(y_test, y_pred_knn_test, zero_division=0)

print(f"\n📊 KNN Results:")
print(f"   - Training Accuracy: {knn_train_acc:.4f} ({knn_train_acc*100:.2f}%)")
print(f"   - Test Accuracy: {knn_test_acc:.4f} ({knn_test_acc*100:.2f}%)")
print(f"   - Precision: {knn_precision:.4f}")
print(f"   - Recall: {knn_recall:.4f}")
print(f"   - F1-Score: {knn_f1:.4f}")

cm_knn = confusion_matrix(y_test, y_pred_knn_test)
fpr_knn, tpr_knn, _ = roc_curve(y_test, y_pred_knn_proba)
roc_auc_knn = auc(fpr_knn, tpr_knn)

# Find optimal k
print("\n🔍 Finding optimal K value...")
k_values = range(1, 21)
k_scores = []
for k in k_values:
    knn_temp = KNeighborsClassifier(n_neighbors=k)
    knn_temp.fit(X_train_scaled, y_train)
    k_scores.append(knn_temp.score(X_test_scaled, y_test))

optimal_k = k_values[np.argmax(k_scores)]
print(f"   ✅ Optimal K = {optimal_k} (Accuracy: {max(k_scores):.4f})")

# Visualization 5: KNN
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Step 5: K-Nearest Neighbors Analysis', fontsize=16, fontweight='bold')

# Plot 1: Confusion Matrix
sns.heatmap(cm_knn, annot=True, fmt='d', cmap='Purples', ax=axes[0, 0],
            xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
axes[0, 0].set_title('Confusion Matrix')
axes[0, 0].set_ylabel('True Label')
axes[0, 0].set_xlabel('Predicted Label')

# Plot 2: ROC Curve
axes[0, 1].plot(fpr_knn, tpr_knn, color='purple', lw=2,
                label=f'ROC curve (AUC = {roc_auc_knn:.2f})')
axes[0, 1].plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Classifier')
axes[0, 1].set_xlim([0.0, 1.0])
axes[0, 1].set_ylim([0.0, 1.05])
axes[0, 1].set_xlabel('False Positive Rate')
axes[0, 1].set_ylabel('True Positive Rate')
axes[0, 1].set_title('ROC Curve')
axes[0, 1].legend(loc="lower right")
axes[0, 1].grid(True, alpha=0.3)

# Plot 3: Metrics
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
scores = [knn_test_acc, knn_precision, knn_recall, knn_f1]
bars = axes[1, 0].bar(metrics, scores, color=['#8e44ad', '#9b59b6', '#a569bd', '#bb8fce'])
axes[1, 0].set_title('Performance Metrics')
axes[1, 0].set_ylabel('Score')
axes[1, 0].set_ylim([0, 1.1])
for bar in bars:
    height = bar.get_height()
    axes[1, 0].text(bar.get_x() + bar.get_width()/2., height,
                    f'{height:.3f}', ha='center', va='bottom', fontweight='bold')

# Plot 4: K value optimization
axes[1, 1].plot(k_values, k_scores, marker='o', linewidth=2, markersize=8, color='mediumpurple')
axes[1, 1].set_title('KNN Performance vs K Value')
axes[1, 1].set_xlabel('Number of Neighbors (K)')
axes[1, 1].set_ylabel('Accuracy')
axes[1, 1].grid(True, alpha=0.3)
axes[1, 1].axvline(x=optimal_k, color='red', linestyle='--',
                   label=f'Optimal K = {optimal_k}', linewidth=2)
axes[1, 1].legend()

plt.tight_layout()
plt.savefig('05_knn.png', dpi=300, bbox_inches='tight')
plt.show()

# ============================================================================
# 6. SUPPORT VECTOR MACHINE (SVM)
# ============================================================================
print("\n" + "=" * 80)
print("STEP 6: SUPPORT VECTOR MACHINE (SVM)")
print("=" * 80)

print("\n⚡ Training SVM Model...")
svm_model = SVC(kernel='rbf', random_state=42, probability=True)
svm_model.fit(X_train_scaled, y_train)

# Predictions
y_pred_svm_train = svm_model.predict(X_train_scaled)
y_pred_svm_test = svm_model.predict(X_test_scaled)
y_pred_svm_proba = svm_model.predict_proba(X_test_scaled)[:, 1]

# Evaluation
svm_train_acc = accuracy_score(y_train, y_pred_svm_train)
svm_test_acc = accuracy_score(y_test, y_pred_svm_test)
svm_precision = precision_score(y_test, y_pred_svm_test, zero_division=0)
svm_recall = recall_score(y_test, y_pred_svm_test, zero_division=0)
svm_f1 = f1_score(y_test, y_pred_svm_test, zero_division=0)

print(f"\n📊 SVM Results:")
print(f"   - Training Accuracy: {svm_train_acc:.4f} ({svm_train_acc*100:.2f}%)")
print(f"   - Test Accuracy: {svm_test_acc:.4f} ({svm_test_acc*100:.2f}%)")
print(f"   - Precision: {svm_precision:.4f}")
print(f"   - Recall: {svm_recall:.4f}")
print(f"   - F1-Score: {svm_f1:.4f}")

cm_svm = confusion_matrix(y_test, y_pred_svm_test)
fpr_svm, tpr_svm, _ = roc_curve(y_test, y_pred_svm_proba)
roc_auc_svm = auc(fpr_svm, tpr_svm)

# Visualization 6: SVM
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Step 6: Support Vector Machine Analysis', fontsize=16, fontweight='bold')

# Plot 1: Confusion Matrix
sns.heatmap(cm_svm, annot=True, fmt='d', cmap='Oranges', ax=axes[0, 0],
            xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
axes[0, 0].set_title('Confusion Matrix')
axes[0, 0].set_ylabel('True Label')
axes[0, 0].set_xlabel('Predicted Label')

# Plot 2: ROC Curve
axes[0, 1].plot(fpr_svm, tpr_svm, color='orangered', lw=2,
                label=f'ROC curve (AUC = {roc_auc_svm:.2f})')
axes[0, 1].plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Classifier')
axes[0, 1].set_xlim([0.0, 1.0])
axes[0, 1].set_ylim([0.0, 1.05])
axes[0, 1].set_xlabel('False Positive Rate')
axes[0, 1].set_ylabel('True Positive Rate')
axes[0, 1].set_title('ROC Curve')
axes[0, 1].legend(loc="lower right")
axes[0, 1].grid(True, alpha=0.3)

# Plot 3: Metrics
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
scores = [svm_test_acc, svm_precision, svm_recall, svm_f1]
bars = axes[1, 0].bar(metrics, scores, color=['#e67e22', '#d35400', '#e74c3c', '#c0392b'])
axes[1, 0].set_title('Performance Metrics')
axes[1, 0].set_ylabel('Score')
axes[1, 0].set_ylim([0, 1.1])
for bar in bars:
    height = bar.get_height()
    axes[1, 0].text(bar.get_x() + bar.get_width()/2., height,
                    f'{height:.3f}', ha='center', va='bottom', fontweight='bold')

# Plot 4: Support Vectors Info
axes[1, 1].bar(['Support Vectors', 'Total Samples'],
               [svm_model.n_support_.sum(), len(X_train)],
               color=['orangered', 'lightcoral'])
axes[1, 1].set_title('Support Vectors vs Total Training Samples')
axes[1, 1].set_ylabel('Count')
for i, v in enumerate([svm_model.n_support_.sum(), len(X_train)]):
    axes[1, 1].text(i, v + 50, f'{v:,}', ha='center', fontweight='bold')

plt.tight_layout()
plt.savefig('06_svm.png', dpi=300, bbox_inches='tight')
plt.show()

# ============================================================================
# 7. NEURAL NETWORK
# ============================================================================
print("\n" + "=" * 80)
print("STEP 7: NEURAL NETWORK (MULTI-LAYER PERCEPTRON)")
print("=" * 80)

print("\n🧠 Training Neural Network Model...")
nn_model = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=500, random_state=42)
nn_model.fit(X_train_scaled, y_train)

# Predictions
y_pred_nn_train = nn_model.predict(X_train_scaled)
y_pred_nn_test = nn_model.predict(X_test_scaled)
y_pred_nn_proba = nn_model.predict_proba(X_test_scaled)[:, 1]

# Evaluation
nn_train_acc = accuracy_score(y_train, y_pred_nn_train)
nn_test_acc = accuracy_score(y_test, y_pred_nn_test)
nn_precision = precision_score(y_test, y_pred_nn_test, zero_division=0)
nn_recall = recall_score(y_test, y_pred_nn_test, zero_division=0)
nn_f1 = f1_score(y_test, y_pred_nn_test, zero_division=0)

print(f"\n📊 Neural Network Results:")
print(f"   - Training Accuracy: {nn_train_acc:.4f} ({nn_train_acc*100:.2f}%)")
print(f"   - Test Accuracy: {nn_test_acc:.4f} ({nn_test_acc*100:.2f}%)")
print(f"   - Precision: {nn_precision:.4f}")
print(f"   - Recall: {nn_recall:.4f}")
print(f"   - F1-Score: {nn_f1:.4f}")
print(f"   - Number of iterations: {nn_model.n_iter_}")

cm_nn = confusion_matrix(y_test, y_pred_nn_test)
fpr_nn, tpr_nn, _ = roc_curve(y_test, y_pred_nn_proba)
roc_auc_nn = auc(fpr_nn, tpr_nn)

# Visualization 7: Neural Network
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Step 7: Neural Network (MLP) Analysis', fontsize=16, fontweight='bold')

# Plot 1: Confusion Matrix
sns.heatmap(cm_nn, annot=True, fmt='d', cmap='Reds', ax=axes[0, 0],
            xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
axes[0, 0].set_title('Confusion Matrix')
axes[0, 0].set_ylabel('True Label')
axes[0, 0].set_xlabel('Predicted Label')

# Plot 2: ROC Curve
axes[0, 1].plot(fpr_nn, tpr_nn, color='crimson', lw=2,
                label=f'ROC curve (AUC = {roc_auc_nn:.2f})')
axes[0, 1].plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Classifier')
axes[0, 1].set_xlim([0.0, 1.0])
axes[0, 1].set_ylim([0.0, 1.05])
axes[0, 1].set_xlabel('False Positive Rate')
axes[0, 1].set_ylabel('True Positive Rate')
axes[0, 1].set_title('ROC Curve')
axes[0, 1].legend(loc="lower right")
axes[0, 1].grid(True, alpha=0.3)

# Plot 3: Metrics
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
scores = [nn_test_acc, nn_precision, nn_recall, nn_f1]
bars = axes[1, 0].bar(metrics, scores, color=['#c0392b', '#e74c3c', '#ec7063', '#f1948a'])
axes[1, 0].set_title('Performance Metrics')
axes[1, 0].set_ylabel('Score')
axes[1, 0].set_ylim([0, 1.1])
for bar in bars:
    height = bar.get_height()
    axes[1, 0].text(bar.get_x() + bar.get_width()/2., height,
                    f'{height:.3f}', ha='center', va='bottom', fontweight='bold')

# Plot 4: Loss Curve
axes[1, 1].plot(nn_model.loss_curve_, linewidth=2, color='darkred')
axes[1, 1].set_title('Training Loss Curve')
axes[1, 1].set_xlabel('Iteration')
axes[1, 1].set_ylabel('Loss')
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('07_neural_network.png', dpi=300, bbox_inches='tight')
plt.show()

# ============================================================================
# 8. COMPREHENSIVE MODEL COMPARISON
# ============================================================================
print("\n" + "=" * 80)
print("STEP 8: COMPREHENSIVE MODEL COMPARISON")
print("=" * 80)

# Collect all results
models_summary = {
    'Logistic Regression': {
        'accuracy': lr_test_acc,
        'precision': lr_precision,
        'recall': lr_recall,
        'f1': lr_f1,
        'auc': roc_auc_lr,
        'fpr': fpr_lr,
        'tpr': tpr_lr
    },
    'Decision Tree': {
        'accuracy': dt_test_acc,
        'precision': dt_precision,
        'recall': dt_recall,
        'f1': dt_f1,
        'auc': roc_auc_dt,
        'fpr': fpr_dt,
        'tpr': tpr_dt
    },
    'KNN': {
        'accuracy': knn_test_acc,
        'precision': knn_precision,
        'recall': knn_recall,
        'f1': knn_f1,
        'auc': roc_auc_knn,
        'fpr': fpr_knn,
        'tpr': tpr_knn
    },
    'SVM': {
        'accuracy': svm_test_acc,
        'precision': svm_precision,
        'recall': svm_recall,
        'f1': svm_f1,
        'auc': roc_auc_svm,
        'fpr': fpr_svm,
        'tpr': tpr_svm
    },
    'Neural Network': {
        'accuracy': nn_test_acc,
        'precision': nn_precision,
        'recall': nn_recall,
        'f1': nn_f1,
        'auc': roc_auc_nn,
        'fpr': fpr_nn,
        'tpr': tpr_nn
    }
}

print("\n📊 Model Performance Summary:")
print("=" * 80)
comparison_df = pd.DataFrame({
    'Model': list(models_summary.keys()),
    'Accuracy': [models_summary[m]['accuracy'] for m in models_summary.keys()],
    'Precision': [models_summary[m]['precision'] for m in models_summary.keys()],
    'Recall': [models_summary[m]['recall'] for m in models_summary.keys()],
    'F1-Score': [models_summary[m]['f1'] for m in models_summary.keys()],
    'AUC': [models_summary[m]['auc'] for m in models_summary.keys()]
})

print(comparison_df.to_string(index=False))

# Find best model
best_model = comparison_df.loc[comparison_df['Accuracy'].idxmax(), 'Model']
best_accuracy = comparison_df['Accuracy'].max()
print(f"\n🏆 Best Model: {best_model} (Accuracy: {best_accuracy:.4f})")

# Create comprehensive comparison visualization
fig = plt.figure(figsize=(20, 12))
fig.suptitle('Complete Model Comparison - All Metrics', fontsize=18, fontweight='bold', y=0.995)

# Create grid
gs = fig.add_gridspec(3, 3, hspace=0.3, wspace=0.3)

# Plot 1: Accuracy Comparison
ax1 = fig.add_subplot(gs[0, 0])
models = list(models_summary.keys())
accuracies = [models_summary[m]['accuracy'] for m in models]
colors_acc = ['#3498db', '#2ecc71', '#9b59b6', '#e67e22', '#e74c3c']
bars = ax1.bar(models, accuracies, color=colors_acc)
ax1.set_title('Model Accuracy Comparison', fontweight='bold')
ax1.set_ylabel('Accuracy')
ax1.set_ylim([0, 1.1])
ax1.tick_params(axis='x', rotation=45)
for bar in bars:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height,
            f'{height:.3f}', ha='center', va='bottom', fontweight='bold')

# Plot 2: Precision Comparison
ax2 = fig.add_subplot(gs[0, 1])
precisions = [models_summary[m]['precision'] for m in models]
bars = ax2.bar(models, precisions, color=colors_acc)
ax2.set_title('Precision Comparison', fontweight='bold')
ax2.set_ylabel('Precision')
ax2.set_ylim([0, 1.1])
ax2.tick_params(axis='x', rotation=45)
for bar in bars:
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height,
            f'{height:.3f}', ha='center', va='bottom', fontweight='bold')

# Plot 3: Recall Comparison
ax3 = fig.add_subplot(gs[0, 2])
recalls = [models_summary[m]['recall'] for m in models]
bars = ax3.bar(models, recalls, color=colors_acc)
ax3.set_title('Recall Comparison', fontweight='bold')
ax3.set_ylabel('Recall')
ax3.set_ylim([0, 1.1])
ax3.tick_params(axis='x', rotation=45)
for bar in bars:
    height = bar.get_height()
    ax3.text(bar.get_x() + bar.get_width()/2., height,
            f'{height:.3f}', ha='center', va='bottom', fontweight='bold')

# Plot 4: F1-Score Comparison
ax4 = fig.add_subplot(gs[1, 0])
f1_scores = [models_summary[m]['f1'] for m in models]
bars = ax4.bar(models, f1_scores, color=colors_acc)
ax4.set_title('F1-Score Comparison', fontweight='bold')
ax4.set_ylabel('F1-Score')
ax4.set_ylim([0, 1.1])
ax4.tick_params(axis='x', rotation=45)
for bar in bars:
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height,
            f'{height:.3f}', ha='center', va='bottom', fontweight='bold')

# Plot 5: AUC Comparison
ax5 = fig.add_subplot(gs[1, 1])
aucs = [models_summary[m]['auc'] for m in models]
bars = ax5.bar(models, aucs, color=colors_acc)
ax5.set_title('AUC Score Comparison', fontweight='bold')
ax5.set_ylabel('AUC')
ax5.set_ylim([0, 1.1])
ax5.tick_params(axis='x', rotation=45)
for bar in bars:
    height = bar.get_height()
    ax5.text(bar.get_x() + bar.get_width()/2., height,
            f'{height:.3f}', ha='center', va='bottom', fontweight='bold')

# Plot 6: All Metrics Heatmap
ax6 = fig.add_subplot(gs[1, 2])
metrics_matrix = comparison_df.set_index('Model').values
sns.heatmap(metrics_matrix, annot=True, fmt='.3f', cmap='YlGnBu',
            xticklabels=['Accuracy', 'Precision', 'Recall', 'F1', 'AUC'],
            yticklabels=comparison_df['Model'].values, ax=ax6, cbar_kws={'shrink': 0.8})
ax6.set_title('All Metrics Heatmap', fontweight='bold')

# Plot 7: ROC Curves - All Models
ax7 = fig.add_subplot(gs[2, :])
colors_roc = ['darkorange', 'green', 'purple', 'orangered', 'crimson']
for i, (model_name, model_data) in enumerate(models_summary.items()):
    ax7.plot(model_data['fpr'], model_data['tpr'], color=colors_roc[i], lw=2,
            label=f'{model_name} (AUC = {model_data["auc"]:.2f})')

ax7.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Classifier')
ax7.set_xlim([0.0, 1.0])
ax7.set_ylim([0.0, 1.05])
ax7.set_xlabel('False Positive Rate', fontsize=12)
ax7.set_ylabel('True Positive Rate', fontsize=12)
ax7.set_title('ROC Curves - All Models Comparison', fontweight='bold', fontsize=14)
ax7.legend(loc="lower right", fontsize=10)
ax7.grid(True, alpha=0.3)

plt.savefig('08_comprehensive_comparison.png', dpi=300, bbox_inches='tight')
plt.show()

# ============================================================================
# FINAL SUMMARY AND RECOMMENDATIONS
# ============================================================================
print("\n" + "=" * 80)
print("FINAL SUMMARY AND RECOMMENDATIONS")
print("=" * 80)

print(f"\n✅ Analysis Complete!")
print(f"\n📈 Total Models Trained: {len(models_summary)}")
print(f"\n🏆 BEST PERFORMING MODEL: {best_model}")
print(f"   - Test Accuracy: {best_accuracy:.4f} ({best_accuracy*100:.2f}%)")
print(f"   - AUC Score: {comparison_df.loc[comparison_df['Model']==best_model, 'AUC'].values[0]:.4f}")

print("\n📊 Model Rankings by Accuracy:")
ranked = comparison_df.sort_values('Accuracy', ascending=False)
for idx, row in ranked.iterrows():
    print(f"   {idx+1}. {row['Model']}: {row['Accuracy']:.4f}")

print("\n💡 Key Insights:")
print(f"   - Highest Accuracy: {comparison_df['Accuracy'].max():.4f}")
print(f"   - Highest Precision: {comparison_df['Precision'].max():.4f}")
print(f"   - Highest Recall: {comparison_df['Recall'].max():.4f}")
print(f"   - Highest F1-Score: {comparison_df['F1-Score'].max():.4f}")
print(f"   - Highest AUC: {comparison_df['AUC'].max():.4f}")

print("\n📁 Generated Files:")
print("   ✅ 01_data_understanding.png")
print("   ✅ 02_data_preprocessing.png")
print("   ✅ 03_logistic_regression.png")
print("   ✅ 04_decision_tree.png")
print("   ✅ 04b_decision_tree_structure.png  (Actual Tree Visualization)")
print("   ✅ 05_knn.png")
print("   ✅ 06_svm.png")
print("   ✅ 07_neural_network.png")
print("   ✅ 08_comprehensive_comparison.png")

print("\n" + "=" * 80)
print("=" * 80)
```
